```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1.2193,1.1090,1.0259,1.1109, 1.0907, 1.1422]
y =[0.5012,0.5372,0.5229,0.5043, 0.5158, 0.5158]
labels = ['Baseline', 'BERT', 'BLF-Spectral', 'MusicNN', 'ResNet', 'VGG19']

m,b = np.polyfit(x, y, 1)

# calculate regression line and show in a scatterplot, different colors for each label and legend on the side
plt.scatter(x, y, color='blue')
for i, txt in enumerate(labels):
    plt.annotate(txt, (x[i], y[i]))
plt.xlabel('Diversity@10')
plt.ylabel('AvgPopularity')
plt.plot(x, m*np.array(x) + b, color='red')
plt.show()
```
